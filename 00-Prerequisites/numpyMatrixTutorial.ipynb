# Numpy Matrix Tutorial

```python
# Importing Numpy Package 
import numpy as np
```

Let's start with defining matrix. There are multiple ways of creating a matrix using numpy
package. We will see how to define a same matrix using multiple methods.

```python
"""Defining two matricies of shape (2,3) Two Rows and Three Columns 

There are two ways of defining a matrix in python.

This is the more popular way of defining a matrix by creating a 1D array and then 
converting it into a matrix.
"""

print "----This is the first way----"

matA = np.arange(6).reshape(2,3)
matB = np.arange(10,16).reshape(3,2)

print "Printing Matrix A"
print matA
print "Printing Matrix B"
print matB

print "Shape of Matrix A: ",matA.shape
print "Shape of Matrix B: ",matB.shape

"""
The other less popular way of defining a matrix is by calling 
builtin np.matrix function.

There are two ways the function can be called.

First way, by passing the string argument where rows are delimited by ;
"""

print "----This is the second way----"

matA = np.matrix('1 2 3;4 5 6')
matB = np.matrix('10 11;12 13;14 15')

print "Printing Matrix A"
print matA
print "Printing Matrix B"
print matB

print "Shape of Matrix A: ",matA.shape
print "Shape of Matrix B: ",matB.shape


# Second way, by passing nested list of rows
print "----This is the third way----"

matA = np.matrix([[1,2,3],[4,5,6]])
matB = np.matrix([[10,11],[12,13],[14,15]])

print "Printing Matrix A"
print matA
print "Printing Matrix B"
print matB

print "Shape of Matrix A: ",matA.shape
print "Shape of Matrix B: ",matB.shape

print "---------------------"
```

After defining the matrix, manytime there are few information about the matrix you need 
to verify. For example:
    1. Size of the matrix
    2. Shape of the matrix
    3. Dimensions of the matrix
    4. Data type of the matrix formed
    5. Total Bytes consumed by the matrix

```python
# Let's start with how the matrix looks like

print "Matrix A: "
print matA

print "\nSize of the Matrix: ", matA.size

print "\nShape of the Matrix: ",matA.shape

print "\nDimensions of the Matrix: ", matA.ndim

print "\nData Type of the Matrix: ", matA.dtype

print "\nTotal size of the Matrix (In Bytes): ", matA.nbytes
```

As we can create a matrix from a 1D array by reshaping it, we can create 1D array from the matrix.

```python
print "Matrix A: "
print matA

# By Attribute
print "\n1D Array from the matrix (By Attribute): "
print matA.A1

#By Function
print "\n1D Array from the matrix (By Function): "
print matA.getA1()
print matA.flatten()


# We can also create a flat iterator object from the matrix similar to 1D array.

matAFlat = matA.flat
print "\nShowing the type of flat iterator object: ",type(matAFlat)
print "\nPrinting Elements: "
for i in matAFlat:
    print i

print "\nWe can also use ravel() function to convert the matrix to a flattened array"
print matA.ravel()
```

Now let's explore how we can get basic details like:
    1. Max/Min value of the matrix
    2. Index of max/min value in the matrix
    4. Sum of all values of the matrix
    5. Sum of values across rows/columns

```python
print matA

print "\n--Matrix level operations--"
print "\nMax Value of the matrix: ",matA.max()
print "\nMin Value of the matrix: ",matA.min()
print "\nIndex of max value in the matrix: ",matA.argmax()
print "\nIndex of min value in the matrix: ",matA.argmin()
print "\nSum of all elements in the matrix: ",matA.sum()
print "\nStandard Deviation of the matrix: ",matA.std()
print "\nMean of the matrix: ",matA.mean()
print "\nCumulative Sum along column and row: "
print "Along Row"
print matA.cumsum(axis = 0)
print "Along Column"
print matA.cumsum(axis = 1)
print "\nCumulative Product along column and row: "
print "Along Row"
print matA.cumprod(axis = 0)
print "Along Column"
print matA.cumprod(axis = 1)


print "\n--Row/Column/Diagonal level operation--"
print "\nMax Value of the matrix along the column: ",matA.max(axis= 0)
print "\nMin Value of the matrix along the row: ",matA.min(axis = 1)
print "\nIndex of max value in the matrix along the column: ",matA.argmax(axis = 0)
print "\nIndex of min value in the matrix along the row: ",matA.argmin(axis = 1)


"""
Columnwise and rowise sum of elements

Axis = 0 for columns 
Axis = 1 for rows 

Note: Opposite of Pandas 
"""
print "\nPrint Columns Sum: ",matA.sum(axis = 0)
print "\nRows Sum: "
print matA.sum(axis = 1)
```

In this section, we will explore how we can sort elements in the matrix. We can do this
at matrix level as well as at rows/columns level

```python
matC = np.matrix([[4,2,6],[43,12,56]])

print "--Matrix Level Sorting--"
print np.sort(matC)

print "\n--Row/Column Level Sorting--"
print "\nColumn Wise Sorting"
print np.sort(matC,axis = 0)
print "\nRow Wise Sorting"
print np.sort(matC,axis = 1)



# We can also get indexes of the sorted matrix both at matrix level as well as at 
# row/column level. We can use np.argsort() for this. 


# print np.argsort(matC,axis = 1)


# We can get index of all non zero values in the matrix using nonzero function.

# print matC.nonzero()
```

```python
# Matrix Transpose 

# Three Methods of Transposing a matrix in numpy

print "Matrix Transpose"

matATranspose = matA.T
print matATranspose

print "-------"

matATranspose = matA.transpose()
print matATranspose

print "-------"

matATranspose = np.transpose(matA)
print matATranspose

print "-------"
matATranspose = matA.getT()
print matATranspose

print ""
print "Conjucate Transpose of a matrix"

"""
Transpose of a matrix is obtained by rearranging columns into rows, or rows into columns. 
The complex conjugate of a matrix is obtained by replacing each element by its 
complex conjugate (i.e x+iy â‡› x-iy or vice versa). The conjugate transpose is obtained 
by performing both operations on the matrix.
"""

matD = np.matrix([[3+2j,3+3j],[2+3j,5+6j]])
print "\nComplex Matrix"
print matD
print "\nConjucate Transpose of the complex matrix"

# By Attribute
print matD.H
print "----"

# By calling function, this is complex conjuct element-wise operation
print matD.conjugate()

# By calling function, this is complex conjugate of all the elements
print matA.conj()
```
